```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df.drop('Prediction', axis=1))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_original = df.drop('Prediction', axis=1) 
y_original = df['Prediction']
```

```python
X_scaled = scaled_df.drop('Prediction', axis=1) 
y_scaled = scaled_df['Prediction']
```

```python
X_train_original, X_test_original, y_train_original, y_test_original = train_test_split( X_original, y_original, test_size=0.2, random_state=42)
```

```python
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled = train_test_split( X_scaled, y_scaled, test_size=0.2, random_state=42)
```

### Perceptron

```python
from sklearn.linear_model import Perceptron 
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def evaluate_perceptron(X_train, X_test, y_train, y_test):
    perceptron = Perceptron(random_state=42)
    perceptron.fit(X_train, y_train)
    y_pred = perceptron.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }
```

```python
# Example usage with the original data
metrics_original = evaluate_perceptron(X_train_original, X_test_original, y_train_original, y_test_original)

print("Original Data Metrics:")
print(f"Accuracy: {metrics_original['accuracy']:.4f}")
print(f"Precision: {metrics_original['precision']:.4f}")
print(f"Recall: {metrics_original['recall']:.4f}")
print(f"F1-Score: {metrics_original['f1']:.4f}")
```

```python
# Example usage with the scaled data
metrics_scaled = evaluate_perceptron(X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled)

print("Scaled Data Metrics:")
print(f"Accuracy: {metrics_scaled['accuracy']:.4f}")
print(f"Precision: {metrics_scaled['precision']:.4f}")
print(f"Recall: {metrics_scaled['recall']:.4f}")
print(f"F1-Score: {metrics_scaled['f1']:.4f}")
```

### Support Vector Machines

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def evaluate_svm(X_train, X_test, y_train, y_test):
    svm = SVC(random_state=42)
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }
```

```python
metrics_original_svm = evaluate_svm(X_train_original, X_test_original, y_train_original, y_test_original)

print("Original Data SVM Metrics:")
print(f"Accuracy: {metrics_original_svm['accuracy']:.4f}")
print(f"Precision: {metrics_original_svm['precision']:.4f}")
print(f"Recall: {metrics_original_svm['recall']:.4f}")
print(f"F1-Score: {metrics_original_svm['f1']:.4f}")
```

```python
metrics_scaled_svm = evaluate_svm(X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled)

print("Scaled Data SVM Metrics:")
print(f"Accuracy: {metrics_scaled_svm['accuracy']:.4f}")
print(f"Precision: {metrics_scaled_svm['precision']:.4f}")
print(f"Recall: {metrics_scaled_svm['recall']:.4f}")
print(f"F1-Score: {metrics_scaled_svm['f1']:.4f}")
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def evaluate_logistic_regression(X_train, X_test, y_train, y_test):
    logistic_regression = LogisticRegression(random_state=42, max_iter=1000)
    logistic_regression.fit(X_train, y_train)
    y_pred = logistic_regression.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }
```

```python
metrics_original_lr = evaluate_logistic_regression(X_train_original, X_test_original, y_train_original, y_test_original)

print("Original Data Logistic Regression Metrics:")
print(f"Accuracy: {metrics_original_lr['accuracy']:.4f}")
print(f"Precision: {metrics_original_lr['precision']:.4f}")
print(f"Recall: {metrics_original_lr['recall']:.4f}")
print(f"F1-Score: {metrics_original_lr['f1']:.4f}")
```

```python
metrics_scaled_lr = evaluate_logistic_regression(X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled)

print("Scaled Data Logistic Regression Metrics:")
print(f"Accuracy: {metrics_scaled_lr['accuracy']:.4f}")
print(f"Precision: {metrics_scaled_lr['precision']:.4f}")
print(f"Recall: {metrics_scaled_lr['recall']:.4f}")
print(f"F1-Score: {metrics_scaled_lr['f1']:.4f}")
```
