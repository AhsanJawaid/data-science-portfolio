### `break`
Stop the execution of the loop.

```python
for item in dept:
    if item == "Finanace":
        break
    print(item)
```

### `continue`
Continue to the next item without executing the lines occuring after `continue` inside the loop.

```python
for item in my_list:
    if item == 1:
        continue
    print(item)
```

```python
dept = ['HR','Finanace', 'Developers', 'Training', 'R&D']
```

### `enumerate()`
In case you need to also know the index: The enumerate () function adds a counter to an iterable and returns it in the form of an enumerating object. 

```python
for location, val in enumerate(dept):
    print(f"Index: {location}, value: {val}")
```

## Looping dictionaries

```python
employee_count = {'HR':5,'Finanace':4, 'Developers':15, 'Training':8, 'R&D':9}

for dept in employee_count:
    print(dept)
```

```python
info_dict = {'Sohail':24, "MS":"DS", "Salary":5000, "Distance":"4km"}
```

```python
for c, d in info_dict.items():
    print(f"{c} and Values are {d}")
```

```python
for a, b in employee_count.items():
    print(f"{a} has {b} number of employees")
```

```python
info_dict = {'Sohail':24, "MS":"DS", "Salary":5000, "Distance":"4km"}
print(info_dict['Sohail'])
```

```python
my_data = {'Name':'Sohail', 'Age': 24, 'Department':'AI&DS', 'Salary':5.6, 'Trainer':True}

for a in my_data:
    print(a, my_data[a])
```

```python
my_dict = {"name": "Alice", "age": 25, "city": "New York"}

# Iterating over keys
for key in my_dict:
    print(key, my_dict[key])

# Or use .items() to get both key and value
for key, value in my_dict.items():
    print(f"{key}: {value}")
```

## `range()`

```python
for number in range(5):
    print(number)
```

```python
for number in range(2, 5):
    print(number)
```

```python
for number in range(0, 10, 3):  # last one is step
    print(number)
```

```python
count = 0
while count < 5: #True, True, True, True, False
    print(count)
    count += 1 # count=1, 2,3,4, 5
```
