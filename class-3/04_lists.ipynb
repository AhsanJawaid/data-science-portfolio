## `sorted(list)`
While `list.sort()` sorts the list in-place, `sorted(list)` returns a new list and leaves the original untouched:

```python
words = ["@", "$", "!", "list", "of", "words"]
sorted(words)
print(f"{sorted(words)}")
```

```python
words
```

```python
numbers = [8, 1, 6, 5, 10]
sorted_numbers = sorted(numbers)
print(f"{numbers=}, {sorted_numbers=}")
```
