```python
data = {'Age': [20,27,21,21,32,21,34,21,34,23],
       'Income': [25000,31000,21000,15000,75000,24000,80000,23000,65000,30000],
       'Rent': [7741,5489,2142,7965,2984,2316,2935,2339,6039,2377]}
```

```python
data
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
```

```python
df = pd.DataFrame(data)
df
```

```python
# Define feature columns and target column
X = df[['Age', 'Income']]
y = df['Rent']
```

```python
X
```

```python
y
```

```python
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X,y)
```

```python
new_point = pd.DataFrame([[30, 50000]], columns=['Age', 'Income'])
new_point
```

```python
predicted_rent = knn.predict(new_point)
print(predicted_rent)
```
