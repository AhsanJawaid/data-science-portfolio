```python
import spacy
```

```python
# LEMMA IN NLP
nlp = spacy.load("en_core_web_sm")

doc = nlp(u'Hello, Tesla CEO`s Name is Elon Musk. His Net income was $9.4 million compared to the prior year of $2.7 million. Revenue exceeded twelve billion dollars, with a loss of $1b.')
for word in doc:
    print(word.text, "=>", word.lemma_)
```

```python
# NLP Pipeline Lemmitizer

#STEM {Snowball Stemmer}
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer(language='english')
doc = "There are many variations of passages of Lorem Ipsum available, but the majority have suffered alteration in some form, by injected humour, or randomised words which don't look even slightly believable."
for token in doc.split(" "):
    print(token, '=>', stemmer.stem(token))
```

```python
#STEM {Porter Stemmer}
from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()
doc = "There are many variations of passages of Lorem Ipsum available, but the majority have suffered alteration in some form, by injected humour, or randomised words which don't look even slightly believable."
for token in doc.split(" "):
    print(token, '=>', stemmer.stem(token))
```
