## Operator precedence in calculations
Mathematical operator precedence applies. Use brackets if you want to change the execution order:

DMAS
BODMAS

```python
print(1 + 2**2 * 3 / 6)  # 1 + 4 * 3 / 6 == 1 + 12 / 6 == 1 + 2
print((1 + 2**2) * 3 / 6)
```

```python
# Step 1: Define the person's name and base salary
person_name = "Sohail Ahmed"
base_salary = 50000  # Base salary in local currency

# Step 2: Define additional components of the salary
# Let's assume that Sohail gets additional components like bonuses or allowances
bonus_percentage = 0.10  # Bonus is 10% of base salary
allowance = 5000         # Fixed allowance

# Step 3: Calculate the total salary
bonus = base_salary * bonus_percentage
total_salary = base_salary + bonus + allowance

# Step 4: Generate the output message
salary_statement = f"Employee: {person_name}\nBase Salary: {base_salary} \nBonus: {bonus} \nAllowance: {allowance} \nTotal Salary: {total_salary}"

# Step 5: Print the salary details
print(salary_statement)
```
