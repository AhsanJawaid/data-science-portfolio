## `if-elif-else`

```python
val = 8
if val >= 100:
    print("value is equal or greater than 100")
elif val > 7:
    print("value is greater than 7 but less than 100")
else:
    print("value is equal or less than 10")
```

```python
salary = 50000

if salary > 50000:
    salary = salary*1.15
    print('Salary increased upto 15%', {salary})
elif salary > 30000:
    salary = salary*1.2
    print('Salary increased upto 20% ', {salary})
else:
    salary = salary*1.5
    print('Salary increased upto 50%', {salary})
```

```python
a = int(input('Enter Number:'))

if a%2==0:
    print('Even Number')
else:
    print('Odd Number')
```

You can have as many `elif` statements as you need. In addition, `else` at the end is not mandatory.

```python
greeting = "Hello fellow Pythonista!"
language = input('Enter your Language ')

if language == "Swedish":
    greeting = "Hejsan!"
    print(greeting)
elif language == "Finnish":
    greeting = "Latua perkele!"
    print(greeting)
elif language == "Spanish":
    greeting = "Hola!"
    print(greeting)
elif language == "German":
    greeting = "Guten Tag!"
    print(greeting)
else:
    print(greeting)
```
